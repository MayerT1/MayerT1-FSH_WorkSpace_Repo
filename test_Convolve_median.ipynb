```
from scipy import ndimage, misc
import matplotlib.pyplot as plt
import numpy as np
```

```
from scipy import ndimage, misc
import matplotlib.pyplot as plt
fig = plt.figure()
plt.gray()  # show the filtered result in grayscale
ax1 = fig.add_subplot(121)  # left side
ax2 = fig.add_subplot(122)  # right side
ascent = misc.ascent()
result = ndimage.median_filter(ascent, size=20)
ax1.imshow(ascent)
ax2.imshow(result)
plt.show()
```

```
x = np.arange(10000).reshape(100,100)

print(x)
```

```
def movingaverage (values,window):
    weights = np.repeat(1.0, window)/window
    smas = np.convolve(values,weights,'valid')
    return smas

x_ma = movingaverage(x.ravel(),20)
print("convolve_ydata", x_ma)
print("convolve_ydatasize:", x_ma.size)
print("convolve_ydata shape:", x_ma.shape)
```

```

result = ndimage.median_filter(x, size=20)
print("convolve_ydata", result)
print("convolve_ydatasize:", result.size)
print("convolve_ydata shape:", result.shape)
print(result.ravel())
```
